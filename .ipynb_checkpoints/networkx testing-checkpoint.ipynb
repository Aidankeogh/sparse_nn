```python
%matplotlib inline
import networkx as nx
from networkx.drawing.nx_agraph import write_dot, graphviz_layout
import matplotlib.pyplot as plt
import numpy as np
import random
from coolname import generate_slug as name
```

```python
inputs = [name(2) for i in range(5)]
hidden = [name(2) for i in range(10)]
done = []
outputs = [name(2) for i in range(5)]
g = nx.DiGraph()

edges = []
    
for h in hidden:
    for _ in range(3):
        edges.append((random.choice(inputs+done),h,1))
        
    for _ in range(1):
        edges.append((h,random.choice(outputs),1))
        
    done.append(h)

g.add_weighted_edges_from(edges)
```

```python
write_dot(g,'test.dot')
pos = graphviz_layout(g, prog='dot')
nx.draw(g, pos, 
        nodelist=inputs,
        node_color='.1',
        with_labels=False, arrows=True)
nx.draw(g, pos, 
        nodelist=hidden,
        node_color='.4',
        with_labels=False, arrows=True)
nx.draw(g, pos, 
        nodelist=outputs,
        node_color='.6',
        with_labels=False, arrows=True)
plt.savefig('nx_test.png')
```

```python
def get_layers(g):
    g2 = nx.topological_sort(g)
    max_layer = nx.dag_longest_path_length(g)
    for n in g2:
        g.node[n]['layer'] = max([g.node[k[0]]['layer'] for k in g.in_edges(n)] + [-1]) + 1
    layers = [[] for _ in range(max_layer + 1)]
    
    for n in g:
        layers[g.node[n]['layer']].append(n)
        
    for i, layer in enumerate(layers):
        for j, n in enumerate(layer):
            g.node[n]['index'] = (i,j)
    
    print(layers)
    return layers
```
